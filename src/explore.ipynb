```python
import requests
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Define the URL
url = "https://raw.githubusercontent.com/4GeeksAcademy/data-preprocessing-project-tutorial/main/AB_NYC_2019.csv"

# Define the folder where you want to save the file
folder_path = os.path.expanduser("/Users/taylorclements/Desktop/Coding")  # Change to your preferred location
file_path = os.path.join(folder_path, "AB_NYC_2019.csv")

# Create the folder if it doesnâ€™t exist
os.makedirs(folder_path, exist_ok=True)

# Download and save the file
response = requests.get(url)
with open(file_path, "wb") as file:
    file.write(response.content)

print(f"File downloaded successfully and saved as {file_path}")

```

```python
#no duplicates in the name column



fig, axis = plt.subplots(3, 2, figsize=(9, 6))

# Create a multiple histogram
sns.histplot(ax = axis[0, 0], data = df_2, x = "price").set_xlim(0, 1000)   # Set the x-axis limits
sns.histplot(ax = axis[0, 1], data = df_2, x = "minimum_nights").set_xlim(0, 30)
sns.histplot(ax = axis[1, 0], data = df_2, x = "number_of_reviews").set(xlim = (0, 300), ylim=(0,4500))  # Set the limits
sns.histplot(ax = axis[1, 1], data = df_2, x = "reviews_per_month").set_xlim(0, 10)
sns.histplot(ax = axis[2, 0], data = df_2, x = "calculated_host_listings_count").set(xlim = (0, 20), ylim=(0, 40000))
sns.histplot(ax = axis[2, 1], data = df_2, x = "availability_365").set_xlim(0, 365)
axis[2,1].set_yscale("log")
# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
#Let's make a heatmap to see if there are any correlations between the columns

df_2["neighbourhood_group"] = pd.factorize(df_2["neighbourhood_group"])[0]
df_2["room_type"] = pd.factorize(df_2["room_type"])[0]
#df_2["neighbourhood"] = pd.factorize(df_2["neighbourhood"])[0]

fig, axis = plt.subplots(figsize=(20, 16))

sns.heatmap(df_2[["neighbourhood_group", "room_type", "price", "minimum_nights", "number_of_reviews", "reviews_per_month", "calculated_host_listings_count", "availability_365"]].corr(), annot=True, fmt=".2f", cmap="coolwarm", annot_kws={"size": 25})

axis.set_xticklabels(axis.get_xticklabels(), rotation=45, horizontalalignment="right", fontsize=20)
axis.set_yticklabels(axis.get_yticklabels(), fontsize=20)
```

```python
sns.pairplot(data=df_2[["neighbourhood_group", "room_type", "price", "minimum_nights", "number_of_reviews", "reviews_per_month", "calculated_host_listings_count", "availability_365"]])
```

```python
#going to dig a bit deeper into the data now and maybe trim some columns and outliers, etc

df_3 = df_2.drop(columns=["last_review", "reviews_per_month"])

print(df_3.columns)
```

```python
#columns that need cleaning/examination: price, min nights, number of reviews, calculated host listings count, availability 365

df_3.describe()
```

```python
#based on this, let's limit price to 500, min nights to 30, number of reviews to 70, calculated host listings count to 10
#, and availability 365 to 365

#just want to look at calculated host listings

print(df_3["calculated_host_listings_count"].value_counts())
```

```python
#let's delete all rows where calculated host listings count is greater than 5

df_4 = df_3[df_3["calculated_host_listings_count"] <= 5]
df_4 = df_4[df_4["price"] <= 500]
df_4 = df_4[df_4["minimum_nights"] <= 15]
```

```python
#no duplicates in the name column



fig, axis = plt.subplots(3, 2, figsize=(9, 6))

# Create a multiple histogram
sns.histplot(ax = axis[0, 0], data = df_4, x = "price").set_xlim(0, 500)   # Set the x-axis limits
sns.histplot(ax = axis[0, 1], data = df_4, x = "minimum_nights").set_xlim(0, 15)
sns.histplot(ax = axis[1, 0], data = df_4, x = "number_of_reviews").set(xlim = (0, 150), ylim=(0,4500))  # Set the limits
sns.histplot(ax = axis[2, 0], data = df_4, x = "calculated_host_listings_count").set(xlim = (0, 5), ylim=(0, 30000))
sns.histplot(ax = axis[2, 1], data = df_4, x = "availability_365").set_xlim(0, 365)
axis[2,1].set_yscale("log")
# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```
